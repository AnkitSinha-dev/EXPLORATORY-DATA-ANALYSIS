### Here the comparison of literacy rates of urban areas of different states.we slightly see an increase in literacy rates of each state.

```python
lr_stat=lr_gr.get_group('State')
a=lr_stat['Literacy Rate (Persons) - Total - 2001']
b=lr_stat['Literacy Rate (Persons) - Total - 2011']
c=((b-a)/a)*100

lr_stat["increaed perc total 2011"]=c
    
```

```python
lr_stat.head(5)
```

```python
lr_stat.sort_values('increaed perc total 2011',ascending=False).plot(kind='bar',x= 'Country/ States/ Union Territories Name',y='increaed perc total 2011',figsize=(10,10))
```

```python
lr_statee=lr_gr.get_group('State')
xx=lr_stat['Literacy Rate (Persons) - Rural - 2001']
yy=lr_stat['Literacy Rate (Persons) - Rural - 2011']
zz=((yy-xx)/xx)*100
lr_statee["increaed perc total rural"]=zz
    
lr_statee.sort_values('increaed perc total rural',ascending=False).plot(kind='bar',x= 'Country/ States/ Union Territories Name',y='increaed perc total rural',figsize=(10,10))
```

```python
lr_statee=lr_gr.get_group('State')
xxx=lr_stat['Literacy Rate (Persons) - Urban - 2001']
yyy=lr_stat['Literacy Rate (Persons) - Urban - 2011']
zzz=((yyy-xxx)/xxx)*100
lr_statee["increaed perc total urban"]=zzz
    
lr_statee.sort_values('increaed perc total urban',ascending=False).plot(kind='bar',x= 'Country/ States/ Union Territories Name',y='increaed perc total urban',figsize=(10,10))
```

```python
lr_statee.sort_values('increaed perc total urban',ascending=False).plot(kind='line',x= 'Country/ States/ Union Territories Name',y='increaed perc total urban',figsize=(10,10))
```
