## Q.Which State has Highest Suicide in 2001?

```python
df.Year=df.Year.astype('int64')
```

```python
df7=df.loc[df.Year==2001]
a=df7.groupby(['State'])['Total'].sum()
print(a)
print(a.max())
print(a.min())


```

#### So Maharashtra has highest Suicides in year 2001 and Lakswdeep has 0 suicides

```python
## Which year has Most No. of Suicides?
b=df.groupby(['Year'])['Total'].sum()
print(b)
print(b.max())
print(b.min())
```

## So in Year 2011 there is more no. of suicides and in year 2001,there is least

```python
## Which state has most suicides overall between 2001-2012?
c=df.groupby(['State'])['Total'].sum()
print(c)
print(c.max())
print(c.min())
```

## Maharashtra has more Suicides in between 2001-2012 and lakshwadeep has least.

```python
## Which age-group has most number of suicides?
cc=df.groupby(['Age_group'])['Total'].sum()
print(cc)
print(cc.max())
print(cc.min())
```

##### So,in general 30-44 age groups has done more suicides and least is done by age groups 0-14,i think they are childrens but still numbers are high.

```python
## Which state has more population overall?
ccc=df.groupby(['State'])['Total_Population'].sum()
print(ccc)
print(ccc.max())
print(ccc.min())
```

##### So uttar pradesh has more population overall between 2001-2012 and Lakshwadeep has least population.
